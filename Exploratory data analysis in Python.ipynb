# Miles Per Gallon Exploratory Data Analysis

```python

import pandas as pd
from pandas import DataFrame, Series
from __future__ import division
import seaborn as sns



data = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data-original",
                   delim_whitespace = True, header=None,
                   names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration',
                            'model', 'origin', 'car_name'])

data.describe()
data.isnull().sum()
data.describe()
data=data.dropna()
print(data.shape)
```

# Exploratory data analysis 

```python
grp_c=data.groupby(['cylinders'])
```

```python
grp =grp_c['mpg'].mean()
```

```python
print(grp)
```

```python

import seaborn as sns
ax = sns.boxplot(x="cylinders", y="mpg", data=data)
```

```python
hh=sns.countplot(x="cylinders",hue=None, data=data)
```

```python
ast = sns.heatmap(data.corr(), vmin=0, vmax=1)
```

```python
sns.regplot(x="horsepower", y="mpg", data=data, ci=95)
```

```python
sns.regplot(x="weight", y="mpg", data=data, ci=95)
```

```python
sns.regplot(x="acceleration", y="mpg", data=data, ci=95)
```

```python
sns.regplot(x="displacement", y="mpg", data=data, ci=95)
```
